```python
approved_collocs_joined = [' '.join(colloc) for colloc in approved_collocs]

for url, desc in descs.items():
    features_for_url = []
    for colloc in approved_collocs_joined:
        
        if desc.find(colloc) > 0:
            features_for_url.append(items.Feature(name='food_name'))
    items.Recipe.objects(origin_url=url).update(set__features=features_for_url)
```

```python
from sklearn import feature_extraction
from sklearn import linear_model, cluster
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
import random

categories = [
    'is_recipe',
    'is_not_recipe'
]



class TrainInput:
    
    def __init__(self, positive_samples, negative_samples):
        self._positive = positive_samples
        self._negative = negative_samples
        self.target = []
        self.data = []
    
    def process_data(self):
        data = []
        for sample in self._positive:
            data.append({'text': sample, 'target': 1})
            
        for sample in self._negative:
            data.append({'text': sample, 'target': 0})
        
        random.shuffle(data)
        self.data = [item['text'] for item in data]
        self.target = [item['target'] for item in data]
    
     
sentences = items.Sentence.objects.all()
train_data = []
for url, desc in descs.items():
    features = items.Recipe.objects(origin_url=url)[0].features
    feat_str = ' '.join(['x{}x'.format('_'.join(feature.name.split())) for feature in features])
    train_data.append(desc + ' ' + feat_str)

data = TrainInput(train_data, [sentence.content for sentence in sentences])
data.process_data()

pipeline = Pipeline([
    ('vect', feature_extraction.text.CountVectorizer()),
    ('tfidf', feature_extraction.text.TfidfTransformer()),
    ('clf', cluster.KMeans())
])

parameters = {
    'vect__max_df': (0.5, 0.75, 1.0),
    #'vect__max_features': (None, 5000, 10000, 50000),
    'vect__ngram_range': ((1, 1), (1, 2)),  # unigrams or bigrams
    #'tfidf__use_idf': (True, False),
    #'tfidf__norm': ('l1', 'l2'),
    # 'clf__alpha': (0.00001, 0.000001),
    
    #'clf__n_iter': (10, 50, 80),
}

grid = GridSearchCV(pipeline, parameters, n_jobs=-1, verbose=1)
grid.fit(data.data, data.target)
```

```python
import feedparser
results = feedparser.parse("http://www.ynet.co.il/Integration/StoryRss975.xml")
```

```python
from bs4 import BeautifulSoup
from pprint import pprint
test_data = []
for entry in results['entries']:
    soup = BeautifulSoup(entry['summary'])
    collocs_in_entry = []
    entry_text = soup.text
    for colloc in approved_collocs_joined:
        if entry_text.find(colloc) > 0:
            collocs_in_entry.append(colloc)
    feat_str = ' '.join(['xfood_namex' for colloc in collocs_in_entry])
    test_data.append(entry_text + ' ' + feat_str)

res = grid.predict(test_data)
clusters = {}
for t, r in zip(test_data, res):
    if r not in clusters:
        clusters[r] = []
    clusters[r].append(t)
for r in clusters:
    print("Category {}:".format(r))
    for i, t in enumerate(clusters[r]):
        print('{}:{}'.format(i, t))
```
